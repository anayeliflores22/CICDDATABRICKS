## # Rol de DataEngineers

```python
%sql
-- CATALOGO
GRANT ALL PRIVILEGES ON CATALOG proyecto_olist TO `DataEngineers`;

-- PERMISOS ESQUEMAS
GRANT CREATE, USE SCHEMA ON SCHEMA  proyecto_olist.bronze TO `DataEngineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  proyecto_olist.silver TO `DataEngineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  proyecto_olist.golden TO `DataEngineers`;

-- PERMISOS CREACION TABLAS
GRANT CREATE TABLE ON SCHEMA proyecto_olist.bronze TO `DataEngineers`;
GRANT CREATE TABLE ON SCHEMA proyecto_olist.silver TO `DataEngineers`;
GRANT CREATE TABLE ON SCHEMA proyecto_olist.golden TO `DataEngineers`;

-- PERMISOS MODIFICACION TABLAS
GRANT MODIFY ON SCHEMA proyecto_olist.bronze TO `DataEngineers`;
GRANT MODIFY ON SCHEMA proyecto_olist.silver TO `DataEngineers`;
GRANT MODIFY ON SCHEMA proyecto_olist.golden TO `DataEngineers`;

--PERMISOS TABLAS
GRANT SELECT ON TABLE proyecto_olist.bronze.orders TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.bronze.customers TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.bronze.products TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.bronze.order_items TO `DataEngineers`;

GRANT SELECT ON TABLE proyecto_olist.silver.orders TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.silver.customers TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.silver.products TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.silver.order_items TO `DataEngineers`;

GRANT SELECT ON TABLE proyecto_olist.golden.dim_customers TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.golden.dim_products TO `DataEngineers`;
GRANT SELECT ON TABLE proyecto_olist.golden.fact_orders TO `DataEngineers`;
```

```python
%sql
-- PERMISOS EXTERNAL LOCATION
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-catalog-mt` to `DataEngineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-bronze` to `DataEngineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-silver` to `DataEngineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-golden` to `DataEngineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-raw` to `DataEngineers`;
```
