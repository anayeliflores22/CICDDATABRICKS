```python
from pyspark.sql.functions import col, coalesce, lit, current_timestamp
```

```python
dbutils.widgets.text("catalogo", "proyecto_olist")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_silver_products = spark.table(f"{catalogo}.{esquema_source}.products")
```

```python
dim_products = df_silver_products.select(
    col("product_id"),
    coalesce(col("product_category_name"), lit("Sin Categoría")),
    col("product_weight_g"),
    col("product_length_cm"),
    col("product_height_cm"),
    col("product_width_cm")
).withColumn("gold_insert_date", current_timestamp())
```

```python
dim_products.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.dim_products")
```
