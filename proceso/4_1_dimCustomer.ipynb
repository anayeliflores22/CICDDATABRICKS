```python
from pyspark.sql.functions import col, upper, current_timestamp
```

```python
dbutils.widgets.text("catalogo", "proyecto_olist")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_silver_customers = spark.table(f"{catalogo}.{esquema_source}.customers")
```

```python
dim_customers = (
    df_silver_customers
    .select(
        col("customer_id"),
        col("customer_unique_id"),
        upper(col("customer_city")).alias("customer_city"),
        col("customer_state")
    )
    .distinct()
    .withColumn("gold_insert_date", current_timestamp())
)
```

```python
dim_customers.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.dim_customers")
```
