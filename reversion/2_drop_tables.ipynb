## Eliminación de objetos y limpieza de checkpoints.

```python
%python
dbutils.widgets.removeAll()
```

```python
%sql
CREATE WIDGET TEXT storageName default="adlsssmartdataanayeli"
```

```python
%python
storageName = dbutils.widgets.get("storageName")
```

## Tablas Gold

```python
%sql
DROP TABLE IF EXISTS golden.fact_orders;
DROP TABLE IF EXISTS golden.dim_customers;
DROP TABLE IF EXISTS golden.dim_products;
```

## Tablas Silver

```python
%sql
DROP TABLE IF EXISTS silver.orders;
DROP TABLE IF EXISTS silver.customers;
DROP TABLE IF EXISTS silver.order_items;
DROP TABLE IF EXISTS silver.products;
```

## Tablas Bronze

```python
%sql
DROP TABLE IF EXISTS bronze.orders;
DROP TABLE IF EXISTS bronze.customers;
DROP TABLE IF EXISTS bronze.order_items;
DROP TABLE IF EXISTS bronze.products;
```

## Esquemas

```python
%sql
DROP SCHEMA IF EXISTS proyecto_olist.raw;
DROP SCHEMA IF EXISTS proyecto_olist.bronze;
DROP SCHEMA IF EXISTS proyecto_olist.silver;
DROP SCHEMA IF EXISTS proyecto_olist.golden;
```

## External Location

```python
%sql
DROP EXTERNAL LOCATION IF EXISTS `exlt-catalog-mt` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `exlt-raw` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `exlt-bronze` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `exlt-silver` FORCE;
DROP EXTERNAL LOCATION IF EXISTS `exlt-golden` FORCE;
```

## Catalogo

```python
%sql
DROP CATALOG IF EXISTS proyecto_olist CASCADE;
```

## Rutas Físicas

```python
dbutils.fs.rm("abfss://raw@{storageName}.dfs.core.windows.net/_checkpoints/orders/", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/",True)
dbutils.fs.rm(f"abfss://silver@{storageName}.dfs.core.windows.net/",True)
dbutils.fs.rm(f"abfss://golden@{storageName}.dfs.core.windows.net/",True)
```
