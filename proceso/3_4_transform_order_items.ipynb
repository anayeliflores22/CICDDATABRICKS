```python
from pyspark.sql.functions import col, current_timestamp
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "proyecto_olist")
dbutils.widgets.text("esquema_source", "bronze")
dbutils.widgets.text("esquema_sink", "silver")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_items = spark.table(f"{catalogo}.{esquema_source}.order_items")
```

```python
df_items_clean = df_items.select(
    col("order_id"),
    col("order_item_id"),
    col("product_id"),
    col("seller_id"),
    col("shipping_limit_date"),
    col("price").cast("decimal(10,2)"),
    col("freight_value").cast("decimal(10,2)")
).withColumn("silver_insert_date", current_timestamp())
```

```python
df_items_clean.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.order_items")
```
